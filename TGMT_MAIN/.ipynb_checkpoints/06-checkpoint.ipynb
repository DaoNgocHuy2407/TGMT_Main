```python
# Import lớp Image
from PIL import Image
from imgtools import *
```

```python
# Đường dẫn ảnh
my_path = 'C:/Users/ADMIN/OneDrive/Documents/crs/img/01/2.png'
```

```python
img = load_image(my_path)
```

```python
display(img)
```

```python
#Chuyển từ ảnh RGB sang ảnh xám
gray_image = img.convert("L")
display(gray_image)
```

```python
# Phân RGB thành 3 tầng màu khác nhau
red_band, green_band, blue_band = img.split()
display(red_band)
display(green_band)
display(blue_band)
```

```python
# Trộn ảnh
merged_image =  Image.merge("RGB", (red_band, green_band, blue_band))
display(merged_image)
```
